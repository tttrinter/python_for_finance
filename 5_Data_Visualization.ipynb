```python
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
```

## One-Dimensional Data Set

```python
np.random.seed(1000)
y = np.random.standard_normal(20)
x = range(len(y))
plt.plot(x,y)
```

```python
plt.plot(y.cumsum())
plt.grid(True) # adds a grid
plt.axis('tight') # adjusts the axis ranges
```

```python
plt.plot(y.cumsum())
plt.grid(True)
plt.xlim(-1,20)
plt.ylim(np.min(y.cumsum())-1,
        np.max(y.cumsum())+1)
```

```python
plt.figure(figsize=(7,4)) # figsize parameter defines the size of the figure in (width, height)
plt.plot(y.cumsum(),'b',lw=1.5)
plt.plot(y.cumsum(),'ro')
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

## Two-Dimensional Data Set
First generate some data - this will generate a NumPy ndarray of 20 x 2 of standard normally distributed numbers. Then we call the "cumsum" method along the first dimension)

```python
np.random.seed(2000)
y = np.random.standard_normal((20,2)).cumsum(axis=0)
```

```python
plt.figure(figsize=(7,4))
plt.plot(y,lw=1.5) # plots 2 lines
plt.plot(y,'ro') # plots two dotted lines
plt.grid(True)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

```python
plt.figure(figsize = (7,4))
plt.plot(y[:,0], lw=1.5, label='1st')
plt.plot(y[:,1], lw=1.5, label='2nd')
plt.plot(y,'ro')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.ylabel('value')
plt.title('A Simple Plot')
```

### Data on different scales:
There are two approaches to this problem:
* Use a secondary Y-Axis (left and right)
* Create subplots (upper/lower, left/right)

```python
y[:, 0] = y[:, 0] * 100 
plt.figure( figsize =( 7, 4)) 
plt.plot( y[:, 0], lw = 1.5, label =' 1st') 
plt.plot( y[:, 1], lw = 1.5, label =' 2nd') 
plt.plot( y, 'ro') 
plt.grid( True) 
plt.legend( loc = 0) 
plt.axis('tight') 
plt.xlabel('index') 
plt.ylabel('value') 
plt.title('A Simple Plot')
```

### Secondary Y-axis

```python
fig, ax1 = plt.subplots() 
plt.plot(y[:, 0], 'b', lw = 1.5, label =' 1st') 
plt.plot(y[:, 0], 'ro') 
plt.grid(True) 
plt.legend( loc = 8) 
plt.axis('tight') 
plt.xlabel('index') 
plt.ylabel('value 1st') 
plt.title('A Simple Plot') 
ax2 = ax1.twinx() 
plt.plot(y[:, 1], 'g', lw = 1.5, label ='2nd') 
plt.plot(y[:, 1], 'rd') 
plt.legend(loc = 0)
plt.ylabel('value 2nd')
```

This is actually two subplots that overlay each other.

### Separate sub-plots

```python
plt.figure(figsize=(7,5))
plt.subplot(211) # this takes 3 integers numrows, numcols, fignum - they can be separated or not
plt.plot(y[:, 0], lw = 1.5, label =' 1st') 
plt.plot(y[:, 0], 'ro') 
plt.grid(True) 
plt.legend( loc = 8) 
plt.axis('tight') 
plt.ylabel('value') 
plt.title('A Simple Plot') 

plt.subplot(212)
plt.plot(y[:, 1], 'g', lw = 1.5, label ='2nd') 
plt.plot(y[:, 1], 'ro') 
plt.legend(loc = 0)
plt.ylabel('value')
```

## Multiple Plot Types

```python
plt.figure(figsize =( 9, 4)) 
plt.subplot(121) 
plt.plot(y[:, 0], lw = 1.5, label ='1st') 
plt.plot(y[:, 0], 'ro') 
plt.grid( True) 
plt.legend(loc = 0) 
plt.axis('tight') 
plt.xlabel('index') 
plt.ylabel('value') 
plt.title('1st Data Set') 

plt.subplot(122)
plt.bar(np.arange(len( y)), y[:, 1], width = 0.5, color ='g', label ='2nd')
plt.grid(True)
plt.legend(loc=0)
plt.axis('tight')
plt.xlabel('index')
plt.title('2nd Data Set')
```

## Other Plot Styles

### Scatter Plot

```python
y = np.random.standard_normal((1000,2))
plt.figure(figsize=(7,5))
plt.plot(y[:,0], y[:, 1],'ro')
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

```python
plt.figure(figsize=(7,5))
plt.scatter(y[:, 0], y[:,1], marker = 'o')
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

### 3rd Dimension in Color
The Scatter Plot allows for a 3rd dimension to be visualized by color. We'll add random data to show this.

```python
c = np.random.randint(0,10,len(y))
```

```python
plt.figure(figsize=(7,5))
plt.scatter(y[:, 0], y[:,1], c=c, marker = 'o')
plt.colorbar()
plt.grid(True)
plt.xlabel('1st')
plt.ylabel('2nd')
plt.title('Scatter Plot')
```

### Histogram

```python
plt.figure(figsize=(7,5))
plt.hist(y, label=['1st','2nd'],bins=25)
plt.grid(True)
plt.legend(loc=0)
plt.xlabel('value')
plt.ylabel('frequency')
plt.title('Histogram')
```

### Stacked Histogram

```python
plt.figure(figsize =( 7, 4)) 
plt.hist(y,
         label =['1st', '2nd'], 
         color =['b', 'g'], 
         stacked = True, 
         bins = 20) 
plt.grid( True)
plt.legend(loc = 0) 
plt.xlabel('value') 
plt.ylabel('frequency') 
plt.title('Histogram')
```

## Box Plots

```python
fig, ax = plt.subplots(figsize=(7,4))
plt.boxplot(y)
plt.grid(True)
plt.setp(ax, xticklabels=['1st','2nd'])
plt.xlabel('data set')
plt.ylabel('value')
plt.title('Boxplot')
```

```python
from matplotlib.patches import Polygon

# 1. Define the function to be plotted and integrated
def func(x):
    return 0.5 * np.exp(x) + 1

# 2. Set the integral limits
a, b = 0.5, 1.5 # integral limits
x = np.linspace(0,2) # generates a set of x-values for the plot
y = func(x) # evaluate the function for all of the x-values

# 3. Plot the function
fig, ax = plt.subplots(figsize=(7,5))
plt.plot(x, y, 'b', linewidth=2)
plt.ylim(ymin=0)

# 4. Illustrate the integral value, i.e. the area under the function between the lower and upper limits
# Generate the shaded area ("patch") by the use of the Polygon function
Ix = np.linspace(a,b)
Iy = func(Ix)
verts = [(a,0)] + list(zip(Ix,Iy)) + [(b,0)]
poly = Polygon(verts, facecolor='0.7', edgecolor='0.5')
ax.add_patch(poly)

# 5. Add the equation. LaTeX code can be used to make the text pretty
plt.text(0.5*(a+b),1, r"$\int_a^b f(x)\mathrm{d}x$",horizontalalignment='center',fontsize=20)

plt.figtext(0.9, 0.075, '$x$')
plt.figtext(0.075, 0.9, '$f(x)$')

ax.set_xticks((a,b))
ax.set_xticklabels(('$a$','$b$'))
ax.set_yticks([func(a), func(b)])
ax.set_yticklabels(('$f(a)$','$f(b)$'))
plt.grid(True)
```

## Financial Plots
The matplotlib.finance library allows for easy retrieval of historical stock price data from Yahoo!Finance. Looking at data for the German DAX with the ticker: ^GDAXI:


```python
import matplotlib.finance as mpf
```

```python
start = (2014, 5,1)
end = (2014, 6, 30)
ticker = '^GDAXI'
quotes = mpf.quotes_historical_yahoo_ochl(ticker, start, end)

quotes[:2]
```

### Candlestick Plots

```python
fig, ax= plt.subplots(figsize=(8,5))
fig.subplots_adjust(bottom=0.2)
mpf.candlestick_ochl(ax,quotes, width=0.6, colorup='b',colordown='r')
plt.grid(True)
ax.xaxis_date()
ax.autoscale_view()
plt.setp(plt.gca().get_xticklabels(),rotation=30)
```

```python
fig, ax = plt.subplots(figsize=(8,5))
mpf.plot_day_summary_oclh(ax, quotes, colorup='b',colordown='r')
plt.grid(True)
ax.xaxis_date()
plt.title('DAX Index')
plt.ylabel('index level')
plt.setp(plt.gca().get_xticklabels(), rotation=30)
```

## Combined Candlestick and Volume Plot

```python
fig, (ax1, ax2) = plt.subplots(2, sharex = True, figsize =( 8, 6)) 
mpf.candlestick_ochl(ax1, quotes, width = 0.6, colorup ='b', colordown ='r') 
ax1.set_title('Yahoo Inc.') 
ax1.set_ylabel('index level') 
ax1.grid(True) 
ax1.xaxis_date() 
plt.bar(quotes[:, 0] - 0.25, quotes[:, 5], width = 0.5) 
ax2.set_ylabel('volume') 
ax2.grid(True)
ax2.autoscale_view() 
plt.setp(plt.gca().get_xticklabels(), rotation = 30)
```

## 3D Plotting

Volatility surfaces:
* Strikes between 50 and 150
* Time to Maturity between 0.5 and 2.5 years

```python
strike = np.linspace(50,150,24)
ttm = np.linspace(0.5,2.5,24)
strike, ttm = np.meshgrid(strike, ttm)
```

```python
strike[:2]
```

Generating fake implied volatilities:

```python
iv = (strike - 100)**2/(100 * strike)/ttm # fake implied vols
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(9,6))
ax = fig.gca(projection = '3d')

surf = ax.plot_surface(strike, ttm, iv, rstride=2, cstride=2,
                      cmap=plt.cm.coolwarm, linewidth=0.5, 
                      antialiased=True)
ax.set_xlabel('strike')
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')

fig.colorbar(surf,shrink=0.5, aspect=5)
```

```python
fig = plt.figure(figsize =( 8, 5)) 
ax = fig.add_subplot(111, projection ='3d')
ax.view_init(30, 60)
ax.scatter(strike, ttm, iv, zdir ='z', s = 25, c ='b', marker ='^') 
ax.set_xlabel('strike') 
ax.set_ylabel('time-to-maturity')
ax.set_zlabel('implied volatility')
```
